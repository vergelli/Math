# `Conjetura de Collatz`:

### Sea $\tiny C$ una funcion, tal que, $\tiny C: \mathbb{N} \rightarrow \mathbb{N}$ entonces C se define como:

$$
C(n) =
\begin{cases}
\frac{n}{2}, & \text{si }n\text{ es par} \\
3n+1, & \text{si }n\text{ es impar}
\end{cases}
$$

```python
def C( n):
    
    cn = n
    
    while cn != 1:
        
        if cn%2==0:

            cn = (cn/2)
            print( cn)
            
        else:

            cn = (3*cn+1)
            print( cn)
            
    return cn
```

```python
def C_fill( n):
    
    xs = []
    cn = n
    
    while cn != 1:
        
        if cn%2==0:

            cn = (cn/2)
            xs.append( cn)
            
        else:

            cn = (3*cn+1)
            xs.append( cn)
            
    return xs
    
```

```python
def C_save(x):
    
    ys = []
    for i in range(2, x):

        ys.append( C_fill(i))

    return ys
```

```python
# Guarda el maximo valor que toma C(n), para un n determinado

def C_save_max(x):
    
    ys = []
    for i in range(2, x):

        ys.append( max(C_fill(i)))

    return ys
```

```python
# Guarda el cardinal del conjunto de resultados de C(n), para cada iteracion para un n determinado

def C_save_cardinal(x):
    
    ys = []
    for i in range(2, x):

        ys.append( len(C_fill(i)))

    return ys
```

```python
def C_save_sum(x):
    
    ys = []
    for i in range(2, x):

        ys.append( sum(C_fill(i)))

    return ys
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("bmh")


def plot_C_cota( x):
    
    
    Y = C_save_max( x)
    X = [i for i in range(2, x)]
    
    plt.figure(num=1, figsize=(15, 6), dpi=128, facecolor='w', edgecolor='k')
    plt.scatter(X, Y,  marker=".", color = "red")
    plt.title("Cota superior de cada cto. resultante de C(n) por cada iteracion\n")
    plt.xlabel("\nNumero de iteraciones")
    plt.ylabel("Cota superior\n\n", rotation = 90)
    plt.grid(True)
    plt.show()
    
    
```

```python
plot_C_cota(1300)
```
